```python
import kagglehub
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")
print("Path to dataset files:", path)
```

```python
books = pd.read_csv(f"{path}/books.csv")
books.head()
```

```python
books.info()
```

```python
# How many missing values we have
books.isnull().sum()
```

```python
# Category is one of the important features, lets see how it is
books['categories'].value_counts()
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)
plt.ylabel('columns')
plt.show()
```

```python
books['missing_description'] = np.where(books['description'].isna(),1,0)
columns_of_interest = ['num_pages', 'ratings_count', 'average_rating', 'missing_description']
correlation_matrix = books[columns_of_interest].corr(method = "spearman")\

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
```

```python
# Now we would like to know how many books we have we this missing values
books = books.iloc[np.where(~books['description'].isna() & 
                            ~books['num_pages'].isna() &
                            ~books['published_year'].isna() &
                            ~books['average_rating'].isna())]
```

```python
cat_counts = books['categories'].value_counts().reset_index()
cat_counts.columns = ['category', 'count']

# Plot
plt.figure(figsize=(40,6))
sns.barplot(data=cat_counts, y='count', x='category', palette='viridis')
plt.title('Top 20 Book Categories')
plt.ylabel('Count')
plt.xlabel('Category')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
books['words_in_description'] = books['description'].str.split().str.len()
books.head()
```

```python
plt.hist(books['words_in_description'], bins=20)
```

```python
# Okay, so we keep only those books that have more than 25 words in their description
books = books[books['words_in_description'] >= 25]
books.shape
```

```python
# Create a new column to combine the title and subtitle
books.loc[:,"title_and_subtitle"] = np.where(~books['subtitle'].isna() & ~ books['title'].isna(), books['title']+ "_" + books['subtitle'], books['title'])
books.head()
```

```python
# Create a new column as a unique identifier
books.loc[:, "tagged_description"] = books['isbn13'].astype(str) + "_" + books['description']
books.head()
```

```python
books = books.drop(['subtitle', 'missing_description', 'words_in_description'], axis=1)
books.to_csv("cleaned_books_dataset.csv")
```
